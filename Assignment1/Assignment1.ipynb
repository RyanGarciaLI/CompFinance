# Question 1

```python
stock_list = [header for header in stock_data.keys() if type(header) == int]
stock_list
```

### i. Annualized standard deviation of daily return

```python
# Q1(i)
from pandas.core.frame import DataFrame
daily_rtn = {}
for stock in stock_list:
    stock_price = stock_data[stock]
    d_rtn = (stock_price - stock_price.shift(1)) / stock_price.shift(1)
    daily_rtn[stock] = d_rtn.dropna()
daily_rtn = DataFrame(daily_rtn)

annual_std = daily_rtn.std() * np.sqrt(252)
print("Annualized standard deviation of daily return:\n{}".format(annual_std))
```

### ii. Annualized covariance between each pair of stocks

```python
sigma_ab = daily_rtn.cov() * 252
sigma_ab
```

### iii. Correlation coefficient between each pair of stocks.

```python
rho_ab = daily_rtn.corr()
rho_ab
```

## i. With Modern Portfolio Theory

```python
annual_rtn = portfolio["expected 1-yr return"][1:11]
annual_rtn = np.array(annual_rtn)
annual_rtn
```

### i.(a). Find the standard deviation and expected return of an equal weight (EW) portfolio, where all the stocks have a weight wi = 0.1

```python
weights = np.array([0.1] * len(stock_list))

portfolio_rtn = np.sum(weights * annual_rtn)
portfolio_std = np.sqrt(np.dot(weights.T, np.dot(sigma_ab, weights)))
print("portfolio return: {}".format(portfolio_rtn.round(3)))
print("portfolio standard deviation: {}".format(portfolio_std.round(3)))
```

### i.(b). Find the global minimum variance (MV) portfolio. Give the standard deviation and expected return of the portfolio, and the weights of the stocks.

```python
import scipy.optimize as sopt
def portfolio_var(weights):
    return np.dot(weights.T, np.dot(sigma_ab, weights))

# boundaries
bounds = tuple((0,1) for x in range(10))
# constraints
cst = ({'type': 'eq', 'fun':lambda x: np.sum(x) - 1})

optb = sopt.minimize(portfolio_var, weights, method="SLSQP", bounds=bounds, constraints=cst)

portfolio_weights = optb['x']
print("Portfolio weights: {}".format(portfolio_weights.round(3)))

portfolio_std = np.sqrt(portfolio_var(portfolio_weights))
print("Portfolio standard deviation: {}".format(portfolio_std.round(3)))

portfolio_rtn = np.sum(portfolio_weights * annual_rtn)
print("Portfolio return: {}".format(portfolio_rtn.round(3)))
```

### i.(c). Draw the efficient frontier, short sale not allowed

```python
import matplotlib.pyplot as plt

def portfolio_return(weights):
    return np.sum(weights * annual_rtn)

target_rtn = np.linspace(0.09, 0.22, 30)
target_var = []
for rtn in target_rtn:
    constraints = ({"type":"eq", "fun":lambda x:portfolio_return(x) - rtn},
           {"type":"eq", "fun":lambda x:sum(x)-1})
    res = sopt.minimize(portfolio_var, 
                       weights, 
                       method = 'SLSQP', 
                       bounds = bounds, 
                       constraints = constraints)
    target_var.append(res['fun'])

target_var = np.array(target_var)

plt.figure(figsize=(8,4))
plt.scatter(target_var, target_rtn, c=target_rtn/target_var, marker="x")
plt.plot(portfolio_std**2, portfolio_rtn, 'y*', markersize=18)
plt.grid(True)
plt.xlabel("Expected Volatility")
plt.ylabel("Expected Return")
```

### i.(d). Find the optimal portfolio to be combined with the risk free asset, short sale not allowed (corresponds to case (c) above). State the standard deviation and expected return of this portfolio. Give the weights of the stocks.

```python
risk_free = 0.043
def portfolio_slope(weights):
    rtn = np.sum(weights * annual_rtn) - risk_free
    sigma = np.sqrt(np.dot(weights.T, np.dot(sigma_ab, weights)))
    return -rtn / sigma
    

optd = sopt.minimize(portfolio_slope, 
                    weights, 
                    method = 'SLSQP', 
                    bounds = bounds, 
                    constraints = cst)

portfolio_w = optd['x']
print("stock weight in porfolio",portfolio_w.round(3))
portfolio_std = np.sqrt(portfolio_var(optd['x']))
print("porfolio stander diviation", portfolio_std.round(3))
portfolio_rtn = portfolio_return(optd['x'])
print("porfolio return", portfolio_rtn.round(3))
```

### ii.(a). Find the optimal ERC portfolio. State the marginal risk contribution of each stock and give the weights of the stocks.

```python
sigma_ab_list = sigma_ab.values.tolist()
x, y = np.shape(sigma_ab)
def sqp(weights):
    tr = 0
    for i in range(x):
        for j in range(y):
            tr += (np.dot(weights[i], np.dot(sigma_ab_list[i], weights)) - np.dot(weights[j], np.dot(sigma_ab_list[j], weights)))**2
    return tr

cons = ({'type':"eq", 'fun': lambda x: np.sum(x) - 1})
opt_iia = sopt.minimize(sqp, weights, method="SLSQP", bounds=bounds, constraints=cons)

port_w = opt_iia['x']
print("stock weights in portfolio", port_w.round(3))
port_std = np.sqrt(np.dot(np.dot(opt_iia['x'].T, sigma_ab), opt_iia['x']))
port_mri = []
for i in range(x):
    mri = np.dot(sigma_ab_list[i], opt_iia['x']/port_std)
    port_mri.append(mri.round(3))
print("marginal risk contribution", port_mri)
```

### ii.(b). Calculate the leverage factor for the leveraged ERC portfolio, with a standard deviation of the portfolio equal to your answer in 2(i)(a) above. Find the expected return of this portfolio.

```python
portfolio_std = np.sqrt(np.dot(weights.T, np.dot(sigma_ab, weights)))
leverage_factor =  portfolio_std / port_std
print("Leverage factor: {}".format(leverage_factor.round(3)))
exp_rtn = (portfolio_return(port_w) - risk_free) * leverage_factor + risk_free
print("Expected return: {}".format(exp_rtn.round(3)))
```
